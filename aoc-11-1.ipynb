```python
with open('/kaggle/input/aoc-11-fh/input_11.txt', 'r') as file:
    # Read the lines and store them in an array
    grid = file.readlines()
grid = [s.rstrip('\n') for s in grid]
map = [[c for c in line.strip()] for line in grid]
```

```python
index = 0
while index < len(grid):
    if grid[index].replace('.', '') == '':
        # Duplicate and insert the string with only '.' characters at the next index
        grid.insert(index + 1, grid[index])
        index += 2  # Move to the next string
    else:
        index += 1
```

```python
# Transpose the grid to work on columns
transposed_grid = [''.join(row) for row in zip(*grid)]

index = 0
while index < len(transposed_grid):
    if transposed_grid[index].replace('.', '') == '':
        # Duplicate and insert the column with only '.' characters at the next index
        transposed_grid.insert(index + 1, transposed_grid[index])
        index += 2  # Move to the next column
    else:
        index += 1

# Transpose the grid back to its original orientation
grid = [''.join(row) for row in zip(*transposed_grid)]
```

```python
galaxies = []

ongoing_number = 1  # Start the ongoing number

for row_index, row in enumerate(grid):
    for col_index, char in enumerate(row):
        if char == '#':
            position = (row_index, col_index)
            galaxy = {'gNum': ongoing_number, 'positions': position}
            galaxies.append(galaxy)
            ongoing_number += 1  # Increment the ongoing number

# Access galaxy information by ongoing numbers
#for g in galaxies:
#    print(f"Galaxy {g['gNum']} is at position {g['positions']}")
```

```python
distance = []
counter = 1
for g in galaxies:
    for num in range(len(galaxies)-counter):
        x = abs(galaxies[num+counter]['positions'][1] - g['positions'][1])
        y = abs(galaxies[num+counter]['positions'][0] - g['positions'][0])
        distance.append(x+y)
        #print(g, galaxies[num+counter], x, y)
    counter+=1
```

```python
with open('/kaggle/input/aoc-11-fh/input_11.txt', 'r') as file:
    # Read the lines and store them in an array
    grid = file.readlines()
grid = [s.rstrip('\n') for s in grid]
map = [[c for c in line.strip()] for line in grid]
```

```python
galaxies = []

ongoing_number = 1  # Start the ongoing number

for row_index, row in enumerate(grid):
    for col_index, char in enumerate(row):
        if char == '#':
            position = (row_index, col_index)
            galaxy = {'gNum': ongoing_number, 'positions': position}
            galaxies.append(galaxy)
            ongoing_number += 1
```

```python
for index, row in enumerate(grid):
    if row.replace('.', '') == '':
        # If the row contains only '.', replace all characters with 'X'
        grid[index] = 'X' * len(row)
```

```python
transposed_grid = [''.join(row) for row in zip(*grid)]

for index, col in enumerate(transposed_grid):
    if all(char in ['.', 'X'] for char in col):
        # If the column contains only '.', replace all characters with 'X'
        transposed_grid[index] = 'X' * len(col)

# Transpose the grid back to its original orientation
grid = [''.join(row) for row in zip(*transposed_grid)]
```

```python
grid
```

```python
distance = 0
counter = 1
for g in galaxies:
    for num in range(len(galaxies)-counter):
        x = galaxies[num+counter]['positions'][1] - g['positions'][1]
        y = galaxies[num+counter]['positions'][0] - g['positions'][0]
        #print(g, galaxies[num+counter]['positions'], x, y)
        if(x >= 0):
            for stepx in range(x):
                if(grid[g['positions'][0]][g['positions'][1]+stepx+1] == 'X'):
                    distance += 1000000
                    #print('Pos: ', grid[g['positions'][0]][g['positions'][1]+stepx+1])
                else:
                    distance += 1
                    #print('Pos: ', grid[g['positions'][0]][g['positions'][1]+stepx+1])
        else:
            for stepx in range(0, x, -1):
                if(grid[g['positions'][0]][g['positions'][1]+stepx-1] == 'X'):
                    distance += 1000000
                    #print('Neg: ', grid[g['positions'][0]][g['positions'][1]+stepx-1])
                else:
                    distance += 1
                    #print('Neg: ', grid[g['positions'][0]][g['positions'][1]+stepx-1])
        for stepy in range(y):
            if(grid[g['positions'][0]+stepy+1][g['positions'][1]] == 'X'):
                distance += 1000000
            else:
                distance += 1
            #print(grid[g['positions'][0]+stepy+1][g['positions'][1]])
        
    counter+=1
```
