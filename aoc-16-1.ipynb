# **GET DATA**

```python
with open('/kaggle/input/aoc-16-1-fh/aoc_16_input.txt', 'r') as file:
    # Read the lines and store them in an array
    grid = file.readlines()
grid = [s.rstrip('\n') for s in grid]
```

# **PART 1**

```python
directions = {
    "n": (-1, 0, "s"),
    "s": (1, 0, "n"),
    "w": (0, -1, "e"),
    "e": (0, 1, "w"),
}
```

```python
start = [0, 0]
new_paths_ns = []
new_paths_ew = []
energized = grid[:]
```

```python
def get_path(d, start):
    cp = start
    cd = d
    new_x = start[0]
    new_y = start[1]
    while(new_x < len(grid[0]) and new_y < len(grid) and new_x >= 0 and new_y >= 0):
        #print(grid[new_x][new_y], new_x, new_y)
        if(grid[new_x][new_y] == "/"):
            if(cd[2] == 'w'):
                cd = directions["n"]
            elif(cd[2] == 's'):
                cd = directions["e"]
            elif(cd[2] == 'e'):
                cd = directions["s"]
            elif(cd[2] == 'n'):
                cd = directions["w"]
        elif(grid[new_x][new_y] == "\\"):
            if(cd[2] == 'w'):
                cd = directions["s"]
            elif(cd[2] == 's'):
                cd = directions["w"]
            elif(cd[2] == 'e'):
                cd = directions["n"]
            elif(cd[2] == 'n'):
                cd = directions["e"]
        elif(grid[new_x][new_y] == "|" and cd[2] == 'w' or grid[new_x][new_y] == "|" and cd[2] == 'e'):
            cd = directions["s"]
            if([new_x, new_y] in new_paths_ns):
                break
            else:    
                new_paths_ns.append([new_x, new_y])
        elif(grid[new_x][new_y] == "-" and cd[2] == 'n' or grid[new_x][new_y] == "-" and cd[2] == 's'):
            cd = directions["e"]
            if([new_x, new_y] in new_paths_ew):
                break
            else:
                new_paths_ew.append([new_x, new_y])
        modified_string = energized[new_x][:new_y] + '#' + energized[new_x][new_y + 1:]
        energized[new_x] = modified_string
        new_x = cp[0]+int(cd[0])
        new_y = cp[1]+int(cd[1])
        cp = [new_x, new_y]
```

```python
get_path(directions["e"], start)
```

```python
for i in range(10):
    if(new_paths_ns != []):
        #print(new_paths_ns)
        paths_ns = new_paths_ns
        new_paths_ns = []
        for path_ns in paths_ns:
            get_path(directions["n"], path_ns)
    if(new_paths_ew != []):
        #print("EW:", new_paths_ew)
        paths_ew = new_paths_ew
        new_paths_ew = []
        for path_ew in paths_ew:
            get_path(directions["w"], path_ew)
```

```python
start = [0, 0]
new_paths_ns = []
new_paths_ew = []
energized = grid[:]
counter = []

for i in range(len(grid[0])):
    start = [0, i]
    new_paths_ns = []
    new_paths_ew = []
    get_path(directions["s"], start)
    for i in range(10):
        if(new_paths_ns != []):
            #print(new_paths_ns)
            paths_ns = new_paths_ns
            new_paths_ns = []
            for path_ns in paths_ns:
                get_path(directions["n"], path_ns)
        if(new_paths_ew != []):
            #print("EW:", new_paths_ew)
            paths_ew = new_paths_ew
            new_paths_ew = []
            for path_ew in paths_ew:
                get_path(directions["w"], path_ew)
    counter.append(sum(string.count('#') for string in energized))
    energized = grid[:]
    
    
for i in range(len(grid)):
    start = [i, 0]
    new_paths_ns = []
    new_paths_ew = []
    get_path(directions["e"], start)
    for i in range(10):
        if(new_paths_ns != []):
            #print(new_paths_ns)
            paths_ns = new_paths_ns
            new_paths_ns = []
            for path_ns in paths_ns:
                get_path(directions["n"], path_ns)
        if(new_paths_ew != []):
            #print("EW:", new_paths_ew)
            paths_ew = new_paths_ew
            new_paths_ew = []
            for path_ew in paths_ew:
                get_path(directions["w"], path_ew)
    counter.append(sum(string.count('#') for string in energized))
    energized = grid[:]
    
for i in range(len(grid)):
    start = [i, len(grid[0])-1]
    new_paths_ns = []
    new_paths_ew = []
    get_path(directions["w"], start)
    for i in range(10):
        if(new_paths_ns != []):
            #print(new_paths_ns)
            paths_ns = new_paths_ns
            new_paths_ns = []
            for path_ns in paths_ns:
                get_path(directions["n"], path_ns)
        if(new_paths_ew != []):
            #print("EW:", new_paths_ew)
            paths_ew = new_paths_ew
            new_paths_ew = []
            for path_ew in paths_ew:
                get_path(directions["w"], path_ew)
    counter.append(sum(string.count('#') for string in energized))
    energized = grid[:]
    
    
for i in range(len(grid)):
    start = [len(grid)-1, i]
    new_paths_ns = []
    new_paths_ew = []
    get_path(directions["w"], start)
    for i in range(10):
        if(new_paths_ns != []):
            #print(new_paths_ns)
            paths_ns = new_paths_ns
            new_paths_ns = []
            for path_ns in paths_ns:
                get_path(directions["n"], path_ns)
        if(new_paths_ew != []):
            #print("EW:", new_paths_ew)
            paths_ew = new_paths_ew
            new_paths_ew = []
            for path_ew in paths_ew:
                get_path(directions["w"], path_ew)
    counter.append(sum(string.count('#') for string in energized))
    energized = grid[:]
```
