# **GET DATA**

```python
from collections import Counter

with open('/kaggle/input/aoc-7-1-fh/input_7.txt', 'r') as file:
    # Read the lines and store them in an array
    lines = file.readlines()

lines = [s.rstrip('\n') for s in lines]
len(lines)
```

```python
lines = [entry.split() for entry in lines]
lines = [[part[0], int(part[1])] for part in lines]
```

# **PART 1**

```python
def max_second_most_chars(s):
    char_count = Counter(s)
    max_count = max(char_count.values(), default=0)
    
    # Get the characters that appear 'max_count' times
    max_chars = [char for char, count in char_count.items() if count == max_count]
    
    # Remove the characters that appear 'max_count' times
    for char in max_chars:
        del char_count[char]
    
    second_max_count = max(char_count.values(), default=0)
    
    # Get the characters that appear 'second_max_count' times
    second_max_chars = [char for char, count in char_count.items() if count == second_max_count]
    
    return max_count, max_chars, second_max_count, second_max_chars
```

```python
hand_value_sublist = []

for i in range(len(lines)):
    max_count, max_chars, second_max_count, second_max_chars = max_second_most_chars(lines[i][0])
    #print(max_count, max_chars, second_max_count, second_max_chars)
    if(max_count == 5):                             #6
        #print('Five of a kind!', lines[i])
        hand_value_sublist.append(6)
    elif(max_count == 4):                           #5
        #print('Four of a kind!', lines[i])
        hand_value_sublist.append(5)
    elif(max_count == 3 and second_max_count == 2): #4
        #print('Full house', lines[i])
        hand_value_sublist.append(4)
    elif(max_count == 3):                           #3
        #print('Three of a kind!', lines[i]) 
        hand_value_sublist.append(3)
    elif(max_count == 2 and len(max_chars) == 2):   #2
        #print('Two pair!', lines[i]) 
        hand_value_sublist.append(2)
    elif(max_count == 2):                           #1
        #print('One pair!', lines[i])
        hand_value_sublist.append(1)
    elif(max_count == 1):                           #0
        #print('High card!', lines[i])
        hand_value_sublist.append(0)
```

```python
def sort_and_calc(part, lines):
    lines = [sublist + [num] for sublist, num in zip(lines, hand_value_sublist)]
    
    if(part == 1):
        custom_order = {'A': 0, 'K': 1, 'Q': 2, 'J': 3, 'T': 4, '9': 5, '8': 6, '7': 7, '6': 8, '5': 9, '4': 10, '3': 11, '2': 12}
    else:
        custom_order = {'A': 0, 'K': 1, 'Q': 2, 'T': 3, '9': 4, '8': 5, '7': 6, '6': 7, '5': 8, '4': 9, '3': 10, '2': 11, 'J': 12}
    
    sorted_data = sorted(lines, key=lambda x: (-x[2], [custom_order.get(c, ord(c)) for c in x[0]]))
    
    result = 0
    counter = 0
    for line in sorted_data:
        #print(line[1], len(lines)-counter)
        result += line[1] * (len(lines)-counter)
        counter+=1
    return result, sorted_data
```

```python
sort_and_calc(1, lines)
```

```python
def max_second_most_joker_chars(s):
    char_count = Counter(s)
    
    joker_count = s.count('J')
    del char_count['J']

    max_count = max(char_count.values(), default=0)
    
    # Get the characters that appear 'max_count' times
    max_chars = [char for char, count in char_count.items() if count == max_count]
    
    # Remove the characters that appear 'max_count' times
    for char in max_chars:
        del char_count[char]
    
    second_max_count = max(char_count.values(), default=0)
    
    # Get the characters that appear 'second_max_count' times
    second_max_chars = [char for char, count in char_count.items() if count == second_max_count]
    
    return max_count, max_chars, second_max_count, second_max_chars, joker_count
```

```python
hand_value_sublist = []

for i in range(len(lines)):
    max_count, max_chars, second_max_count, second_max_chars, joker_count = max_second_most_joker_chars(lines[i][0])
    #print(max_count, max_chars, second_max_count, second_max_chars, joker_count)
        
    max_count += joker_count    
    
    if(max_count == 5 or joker_count == 5):                             #6
        #print('Five of a kind!', lines[i])
        hand_value_sublist.append(6)
    elif(max_count == 4):                           #5
        #print('Four of a kind!', lines[i])
        hand_value_sublist.append(5)
    elif(max_count == 3 and second_max_count == 2 or max_count == 3 and len(max_chars) == 2): #4
        #print('Full house', lines[i])
        hand_value_sublist.append(4)
    elif(max_count == 3):                           #3
        #print('Three of a kind!', lines[i]) 
        hand_value_sublist.append(3)
    elif(max_count == 2 and len(max_chars) == 2):   #2
        #print('Two pair!', lines[i]) 
        hand_value_sublist.append(2)
    elif(max_count == 2):                           #1
        #print('One pair!', lines[i])
        hand_value_sublist.append(1)
    elif(max_count == 1):                           #0
        #print('High card!', lines[i])
        hand_value_sublist.append(0)
```
